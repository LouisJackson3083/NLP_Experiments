A homograph - is a word that shares the same written form as another word but has a different meaning.
Bank (river bank) - Bank (money bank)
Lead (noun) - Lead (verb)
Wind (noun) - Wind (verb)

A homophone - is a word that sounds the same as another word but will differ in meaning, spelling
Knight - Night
Their - There
Time - Thyme

A polysemeus - is a word that has more than one meaning: 
Play
Right

These types of words can cause problems for NLP applications as the task then becomes understanding one of many contexts the word could be in

Task 5: Build a Naïve Bayes classifier using the 6 training examples given in the table below for
disambiguating senses of “hard”? Here, each example is characterised by bag-of-word features, for
which a 8-word dictionary is used: {difficult, task, duty, control, lawyer, strong, liquor, work}.




```python
dataset = {
    'sense2': ['difficult', 'task', 'lawyer', 'difficult', 'time', 'work', 'duty', 'hard', 'control', 'work'],
    'sense1': ['strong', 'liquor', 'dinner', 'liquor', 'pub', 'work', 'liquor', 'liquor', 'wine'],
}

def naive_bayes(x, y):
    count_word_in_class = 0
    count_class_total_words = 0
    total_words = 0
    for key, item in dataset.items():
        if (key == y):
            count_class_total_words = len(key)
            for word in item:
                if word == x:
                    count_word_in_class += 1
        total_words += len(item)

    return (count_word_in_class + 1) / (count_class_total_words + total_words)


naive_bayes('difficult', 'sense2')
```
